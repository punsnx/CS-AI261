# 1. Setup Dependencies

```python
import pandas as pd
import numpy as np
import random
import os

seed_value=11111
random.seed(seed_value)
np.random.seed(seed_value)
os.environ['PYTHONHASHSEED'] = str(seed_value)
```

# 2. Import Data 
(CarPrice.csv for student id at position 9 as number 3 [6610402230])

```python
df = pd.read_csv('CarPrice.csv')
```

# Clean Data

```python
df.dropna(axis='index', inplace=True)
df.info()
```

```python
df
```

```python

y = np.array(df.loc[:, (df.columns == 'ราคาขายมือสอง')].astype(int))
```

```python
categorical_columns = ['ยี่ห้อ', 'ประเภทเชื้อเพลิง', 'ประเภทผู้ขายรถ', 'ประเภทเกียร์']
numerical_columns = ['ปีที่ออกขายครั้งแรก', 'กิโลเมตรที่ขับไปแล้ว', 'จำนวนเจ้าของรถในอดีต']
```

```python
from sklearn.preprocessing import OneHotEncoder

encoders = {}
X = np.empty((df.shape[0], 0))
for col in categorical_columns:
    enc = OneHotEncoder()
    transformed = enc.fit_transform(df[[col]])
    X = np.hstack((X, transformed.toarray()))
    encoders[col] = enc
    
for col in numerical_columns:
    X = np.hstack((X, np.array(df.loc[:, [col]])))
```

```python
X.shape
```

```python
X[0]
```

```python
for enc in encoders.values():
    print(enc.get_feature_names_out()) 
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=seed_value)
```

```python
from sklearn.tree import DecisionTreeRegressor

regressor = DecisionTreeRegressor(random_state=seed_value,criterion = "squared_error",max_depth=5,max_leaf_nodes=11)
regressor.fit(X_train, y_train)
```

```python
regressor.score(X_test, y_test)
```

```python
y_pred = regressor.predict(X_test)
```

```python
import pandas as pd

decoded_df = pd.DataFrame()

start_col = 0
for col, enc in encoders.items():
    end_col = start_col + len(enc.get_feature_names_out())
    decoded_col = enc.inverse_transform(X_test[:, start_col:end_col])
    decoded_df[col] = decoded_col.flatten()
    start_col = end_col

# Decode numerical columns
for i, col in enumerate(numerical_columns):
    decoded_df[col] = X_test[:, start_col + i]
```

```python
test = pd.concat([decoded_df, pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})], axis=1)
test
```

```python
from sklearn.metrics import  mean_squared_error

# Calculate metrics
mse = mean_squared_error(y_test, y_pred)

# Print the report
print(f"Mean Squared Error (MSE): {mse}")
```

```python
from sklearn.tree import export_graphviz

# Flatten the list of feature names from encoders
feature_names = [feature for enc in encoders.values() for feature in enc.get_feature_names_out()]
feature_names += numerical_columns
```

```python
export_graphviz(regressor, out_file='tree.dot', feature_names=feature_names)
```
