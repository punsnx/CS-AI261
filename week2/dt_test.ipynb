```python
X = df.drop(columns=['ราคาขายมือสอง'])
Y = df['ราคาขายมือสอง']
X.shape, Y.shape
```

```python
X = X.to_numpy()
Y = Y.to_numpy()
Y = Y.astype(np.float32)
print(X.shape,Y.shape)
X,Y
```

```python
X[:,0:1]
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = []
for i,idx in enumerate((0,3,4,5)):
    encoder.append(OneHotEncoder())
    encoder[i].fit(X[:,idx:idx+1])
    print(encoder[i].categories_)
```

```python
X_enc = encoder[0].transform(X[:,0:1])
print(X_enc.toarray())
```

```python
X_enc = np.empty((X.shape[0],0))
print(X_enc.shape)
encoders = []
for i in range(X.shape[1]):
    if i in (0,3,4,5):
        encoder = OneHotEncoder()
        encoder.fit(X[:,i:i+1])
        X_enc = np.hstack((X_enc, encoder.transform(X[:,i:i+1]).toarray().astype(np.float32) ))
        encoders.append(encoder)
    else:
        X_enc = np.hstack((X_enc, X[:,i:i+1].astype(np.float32)))

X_enc.shape, X_enc
```

```python
X_enc[0]
```

```python
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
depths = range(1,100)
train_errors = []
test_errorrs = []

X_train,X_test,Y_train,Y_test = train_test_split(X_enc,Y,test_size=0.3,random_state=seed_value)

for depth in depths:
    model = DecisionTreeRegressor(max_depth=depth)
    model.fit(X_train,Y_train)
    y_train_pred = model.predict(X_train)
    train_error = mean_absolute_error(Y_train, y_train_pred)
    train_errors.append(train_error)
    y_test_pred = model.predict(X_test)
    test_errorr = mean_absolute_error(Y_test,y_test_pred)
    test_errorrs.append(test_errorr)
```

```python
# Plotting the error vs max_depth
plt.figure(figsize=(12, 6))
plt.plot(depths, train_errors, marker='o', linestyle='-', color='b', label='Train Error')
plt.plot(depths, test_errorrs, marker='o', linestyle='-', color='r', label='Test Error')
plt.title('Model Performance vs Max Depth', fontsize=14)
plt.xlabel('Max Depth', fontsize=12)
plt.ylabel('Mean Absolute Error (MAE)', fontsize=12)
plt.legend()
plt.show()
```

```python
seeds = range(1000)
train_errors = []
test_errorrs = []

for seed_value in seeds:
    random.seed(seed_value)
    np.random.seed(seed_value)
    os.environ['PYTHONHASHSEED'] = str(seed_value)
    X_train,X_test,Y_train,Y_test = train_test_split(X_enc,Y,test_size=0.3,random_state=seed_value)
    
    model = DecisionTreeRegressor(max_depth=20)
    model.fit(X_train,Y_train)
    y_train_pred = model.predict(X_train)
    train_error = mean_absolute_error(Y_train, y_train_pred)
    train_errors.append(train_error)
    y_test_pred = model.predict(X_test)
    test_errorr = mean_absolute_error(Y_test,y_test_pred)
    test_errorrs.append(test_errorr)
```

```python
# Plotting the error vs max_depth
plt.figure(figsize=(30, 4))
plt.plot(seeds, train_errors, marker='o', linestyle='-', color='b', label='Train Error')
plt.plot(seeds, test_errorrs, marker='o', linestyle='-', color='r', label='Test Error')
plt.title('Model Performance vs Max Depth', fontsize=14)
plt.xlabel('Seed', fontsize=12)
plt.ylabel('Mean Squared Error (MAE)', fontsize=12)
plt.legend()
plt.show()
```
